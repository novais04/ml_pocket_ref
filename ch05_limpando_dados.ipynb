## Janitor (pyjanitor)

```python
import janitor as jn
```

```python
df_janitor = (
    pd.read_excel("dirty_data.xlsx", engine='openpyxl')
   .remove_empty() # Removing empty rows and columns
   .clean_names(strip_underscores=True) #Cleaning columns name (multi-indexes are possible!)
##   .coalesce(['certification', 'certification_1']) #Coalesce multiple columns into a single column
   .convert_excel_date("hire_date") #te conversions (from matlab, excel, unix) to Python datetime format   .rename_column("%_allocated","percent_allocated") #renomeia uma coluna
   .add_column("gratitude_points", gratitude_points) #Add a column to the dataframe.
   .transform_column("gratitude_points", np.log10, "gratitude_log") #Transform the given column in-place using the provided function.
)
df_janitor
```
